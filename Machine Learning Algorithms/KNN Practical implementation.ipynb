```python
import pandas as pd
import numpy as np
import seaborn as sns

pd.set_option('display.max_columns', 60)
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
df=df[df['species']!='virginica']
```

```python
df['species']=df['species'].map({'setosa':0,'versicolor':1})
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

### Standardize the variables

Because the KNN classfier predicts the class of a given test by identifying the observations that are nearest to it, the SCALE of the variables matters. Any variables that are on large scale will have a much larger effect on the distance between the observations , and hence on the KNN clssifer than variables that are on small scale. 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
knn_predict=knn.predict(X_test)
```

### Evaluations

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,knn_predict)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
print(classification_report(y_test,knn_predict))
```
